Yêu cầu 1 — Đọc & Làm sạch dữ liệu (20 điểm)
1. Đọc đủ 3 sheet bằng pandas.
2. Chuẩn hóa tên tỉnh/thành, tên quận/huyện (ví dụ: in hoa/thường, loại bỏ khoảng trắng
thừa).
3. Kiểm tra giá trị null, trùng lặp trên toàn bộ dữ liệu.
4. Chuyển kiểu dữ liệu (Type Casting):
– Ngay → datetime
– GiaBan & SoLuong → int

```python
import pandas as pd

source= "data/data_CuaHangTienLoi_ShopKhai.xlsx"

shop = pd.read_excel(source,sheet_name="CuaHang")
products = pd.read_excel(source,sheet_name="HangHoa")
transactions = pd.read_excel(source,sheet_name="GiaoDich")

shop["TinhThanh"]=shop["TinhThanh"].astype("string")
shop["MaCuaHang"]=shop["MaCuaHang"].astype("string")
shop["TenCuaHang"]=shop["TenCuaHang"].astype("string")
shop["QuanHuyen"]=shop["QuanHuyen"].astype("string")
shop["ViTri"]=shop["ViTri"].astype("string")
shop.set_index("MaCuaHang", inplace=True)

# shop.head()
# def upperCase(ma:str):
#     return ma.title()

shop["TinhThanh"]=shop["TinhThanh"].str.upper("string")
shop["TinhThanh"]=shop["TinhThanh"].str.strip("string")

shop["QuanHuyen"]=shop["QuanHuyen"].str.title("string")
shop["QuanHuyen"]=shop["QuanHuyen"].str.strip("string")


shop.dropna()

transactions["Ngay"]= transactions["Ngay"].astype("datatimes64[ms]")
transactions["Ngay"]= transactions["Ngay"].astype("int64")

products["GiaBan"]= products["GiaBan"].astype("int64")

shop.dropna()
products.dropna()
transactions.dropna()
```

### Yêu cầu 2 — Gộp dữ liệu (15 điểm)
Gộp dữ liệu theo 3 bảng để tạo ra bảng phân tích tổng hợp df_all: GiaoDich → gộp với HangHoa theo MaHang → gộp tiếp với CuaHang theo MaCuaHang

```python
transactions.set_index("MaHang", inplace=True)
df_transactions_product= pd.merge(transactions,products,
                                  on= "MaHang", how="inner")
df_transactions_product.set_index("MaCuaHang", inplace=True)
df_all = pd.merge(df_transactions_product,shop, on="MaCuaHang", how="inner")
df_all
```
